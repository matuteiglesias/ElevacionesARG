## Cargar geometria de radios censales

```python
import geopandas as gpd
import os

if not os.path.exists('./../IGN_INDEC_georef/'):
    !git clone https://github.com/matuteiglesias/IGN_INDEC_georef.git ./../../

radios = gpd.read_file('./../IGN_INDEC_georef/censos_shp_CONICET/2010_RADIOS ARGENTINA')
radios = radios.to_crs(4326)
radios.shape
```

```python
centroides = radios.copy()
centroides['geometry'] = radios.centroid
```

### Loop para pedir elevaciones de radios.

```python
import pandas as pd
```
